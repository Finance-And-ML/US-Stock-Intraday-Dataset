```python
import os
import csv
import datetime
                   
rootpath = './20170718_20170721_1minute/nyse_2705_170718_170721/'   
new_path = './20170718_20170721_1minute/fullminutes_nyse/'

date_list = [(7,18),(7,19),(7,20),(7,21)]

allcsvs = os.listdir(rootpath)

def write_row(csv_writer, row):
    csv_writer.writerow(row)

for csvfilename in allcsvs:
    # only deal with file that has content:
    if os.stat(rootpath+csvfilename).st_size != 0:
        with open(rootpath+csvfilename) as csvfile, open(new_path+csvfilename, 'w') as newcsvfile:
                print("working on "+csvfilename+"...")

                new_ticker_csv = csv.writer(newcsvfile, delimiter=',') 

                ticker_rows = csv.reader(csvfile, delimiter=',')

                # check if the file is empty


                ticker_set = {}

                start_of_a_day = True

                # add all existing minutes to the dictionary:
                for row in ticker_rows:
                    row_date = row[1]
                    row_time = row[2]
                    key = row_date + ' ' + row_time
                    ticker_set.update({row[1]+' '+row[2]:(row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7])})

                 # iterate July 3 to 17 9:30 to 4:00pm to get the by-minute time keys
                keys=[]
                for aDay in date_list:
                    now = datetime.datetime(2017, aDay[0], aDay[1], 9, 30)
                    end = now + datetime.timedelta(minutes=390)
                    while now <= end:
                        keys.append(now.strftime("%Y-%m-%d %H:%M:%S"))
                        now += datetime.timedelta(minutes=1)

                # fill out missing-key values
                previousDateKey = list(ticker_set.keys())[0]
                for dateKey in keys:
                    if dateKey not in ticker_set:
                        previousDateValue = ticker_set.get(previousDateKey)
                        currentDate = dateKey.split(" ")[0]
                        currentTime = dateKey.split(" ")[1]
                        ticker_set.update({dateKey:(previousDateValue[0],currentDate,currentTime,previousDateValue[3],previousDateValue[4],previousDateValue[5],previousDateValue[6],previousDateValue[7],)})
                    else:
                        previousDateKey = dateKey

                for j in sorted(list(ticker_set.keys())):
                    i = ticker_set.get(j)
                    new_ticker_csv.writerow([i[0],i[1],i[2],i[3],i[4],i[5],i[6],i[7]])
```
